```python
#Importing necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
#Read the Dataset of Sample Superstore
df = pd.read_csv("C:/Users/ANSHUMAN PRATIK/Desktop/SampleSuperstore.csv")
df.head()
```

```python
#Locating any missing value in Dataframe
df.isnull().sum()
```

```python
#Data Types of each column (They are Object, Intger and floating point type)
df.dtypes
```

```python
#Removing duplicate records
df.drop_duplicates()
```

```python
#Number of Unique values in each Column
df.nunique()
```

```python
#Obtaining Columns available
df.columns
```

```python
#'Correlation' between Columns variables [Sales and Profit is highly correlated- nearly50% correlation]
df.corr()
```

```python
#'Covariance' observed [Sales and Profit quantity observes maximum Covariance- high relationship]
df.cov()
```

```python
#Count of Unique values in 'Region' 
df['Region'].value_counts()
```

```python
#Visualization of Counts of Sales across Region
sb.set_style('whitegrid')
sb.countplot(df['Region'],order=df['Region'].value_counts().index)
```

```python
#Count of Unique values in 'State' 
df['State'].value_counts()
```

```python
#Visualizing Count of Sales across state as Bar Graph
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sb.set_style('whitegrid')
sb.countplot(df['State'], order=df['State'].value_counts().index)
```

```python
#Count of Uniques values of 'City' and displaying Top 5 cities
city=df['City'].value_counts()
city.head()
```

```python
#Bar plot of Sales Count vs 'Category'
sb.set_style('whitegrid')
sb.countplot(x=df['Category'], order=df['Category'].value_counts().index)
```

```python
#Visualizing Count of Sales across 'Sub-Category' as Bar Graph
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sb.set_style('whitegrid')
sb.countplot(df['Sub-Category'], order=df['Sub-Category'].value_counts().index)
```

```python
#Bar plot of Sales Count and 'Customer Segments'
sb.set_style('whitegrid')
sb.countplot(df['Segment'])
```

```python
#Bar Plot of Count of Sales for each 'Sub-Catgeory'
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sb.set_style('whitegrid')
sb.barplot(x=df['Sub-Category'], y=df['Sales'])
```

```python
#Bar Plot showing Profit generated in each 'State'
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sb.set_style('whitegrid')
sb.barplot(x=df['State'], y=df['Profit'])
```

```python
#Graph depicting Amount of Sales from Top Cities
topCity=df.groupby('City')['Sales'].sum().sort_values(ascending =False).iloc[0:10]
plt.title('Sales of Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Total amount of Sales')
topCity.plot(kind='bar')
```

```python
#Graph depicting Number of Sales from Top States
topState=df.groupby('State')['Quantity'].sum().sort_values(ascending =False).iloc[0:10]
plt.title('Sales of Top 10 States')
plt.xlabel('State')
plt.ylabel('Total amount of Sales')
topState.plot(kind='bar')
```

```python
#Grouping of City based on Discount Slabs
disc = df.groupby('City')['Discount'].value_counts().sort_values(ascending =False).iloc[0:10]
disc
```

```python
#Bar Chart of 'Sales' and 'Profit' in each State [Legend: Blue-> Profit & Orange-> Sales]
chart=df.groupby('State')['Profit','Sales'].sum()
chart.plot(kind='bar',figsize=(25,8))
```
