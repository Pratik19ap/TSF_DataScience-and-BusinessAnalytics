The model involves **Machine Learning** algorithm to implement **Linear regression** function using Scikit-Learn library of Python. 


```python
#Importing necessary Libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
import seaborn as sb
```

```python
#Creating Student Data frame and display first Records
data="http://bit.ly/w-data"
student_df = pd.read_csv(data)
student_df.head()
```

```python
#Scatter plot to visualize the relationship between Percentage and Hours studied.
student_df.plot(x='Hours', y='Scores', style='o')  
plt.title('Percentage and Hours studied')  
plt.xlabel('Number of Hours studied')  
plt.ylabel('Percentage Obtained')  
plt.show()
```

```python
#Splitting of dataset consisting of attributes and values into training and testing sets using Scikit-learn library.
x = student_df.iloc[:, :-1].values  
y = student_df.iloc[:, 1].values  

from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0) 
```

```python
#Training the model for Linear Regression function
from sklearn.linear_model import LinearRegression  
reg = LinearRegression()  
reg.fit(x_train, y_train) 
```

```python
#Plotting the Regression line of chracteristic
reg_line = reg.coef_*x+reg.intercept_
plt.plot(reg_line)
plt.title('Characteristic of Percentage with Study Hours')
plt.show()
plt.scatter(x, y)
plt.title('Regression Line')
plt.plot(x, reg_line);
plt.show()
```

```python
#Predicting Scores with Testing data of Study Hours
print(x_test)
y_pred = reg.predict(x_test) 
y_pred
```

```python
#Comparing Actual and Predicted 'Percentage scores'
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
#Estimating Percentage for 9.25 hours of study per day
hours = 9.25
pred_percentage=reg.coef_*hours+reg.intercept_
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(pred_percentage))
```

```python
#Evaluation of model through 'Mean Square error' between Test and Predicted Values
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
```

   **CONCLUSION**
   
• The **positive Linear Correlation** between the Percentage and Study hours was observed and plotted.

• We were able to predict the Percentage Score of student based on his Study hours through **Training and analysis** of model.

• A Student studying for 9.25 hours per day is expected to obtain **93.69%** Score.
